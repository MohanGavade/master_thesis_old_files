```python
import pandas as pd
import pickle
import numpy as np

def process_labels(file_path, required_columns):
    """
    Process the labels pickle file to keep only specific columns
    and convert their values to numeric format.
    Args:
        file_path: Path to the pickle file.
        required_columns: List of columns to retain and process.
    Returns:
        Processed DataFrame with only the required columns in numeric format.
    """
    # Step 1: Load the pickle file
    with open(file_path, "rb") as f:
        labels_df = pickle.load(f)
    
    # Convert to a pandas DataFrame (if it's not already)
    if not isinstance(labels_df, pd.DataFrame):
        labels_df = pd.DataFrame(labels_df)
    
    # Step 2: Drop unwanted columns
    labels_df = labels_df[required_columns]

    # Step 3: Convert columns to numeric format
    def convert_to_float(value):
        """Convert values to float, handling lists, arrays, and scalars."""
        if isinstance(value, list):
            return [float(v) for v in value]
        elif isinstance(value, np.ndarray):
            return [float(v) for v in value.tolist()]
        elif isinstance(value, (int, float, str)):
            try:
                return float(value)
            except ValueError:
                return 0.0  # Default for invalid strings
        else:
            return 0.0  # Default for unsupported types

    # Debugging to identify problematic rows
    for col in required_columns:
        for i, value in enumerate(labels_df[col]):
            try:
                labels_df.at[i, col] = convert_to_float(value)
            except Exception as e:
                print(f"Error converting row {i}, column {col}: {value}")
                print(f"Exception: {e}")
                labels_df.at[i, col] = 0.0  # Set default for problematic rows

    return labels_df

# Define the file path and required columns
file_path = "D:\\temp\\frr40_objects_0.p"
required_columns = [
    "orientation", "x", "y", "width_edge_mean", "length_edge_mean",
    "header.origin.x", "header.origin.y", "header.origin.z",
    "header.origin.roll", "header.origin.pitch", "header.origin.yaw"
]

# Process the labels
processed_labels = process_labels(file_path, required_columns)

# Inspect the processed DataFrame
print("Processed Labels DataFrame:")
print(processed_labels.head())
```

```python
import numpy as np
import pandas as pd

def transform_to_scs_dynamic_list(processed_labels):
    """
    Transform coordinates from Vehicle Coordinate System (VCS) to Sensor Coordinate System (SCS),
    using yaw and ignoring roll, pitch, and z (as they are always zero). Handles header values as uniform offsets.
    """
    # Initialize lists to store transformed results
    x_scs_all, y_scs_all = [], []

    # Iterate through rows
    for i, row in processed_labels.iterrows():
        # Extract yaw and translation offsets
        try:
            yaw = float(row["header.origin.yaw"])
        except ValueError:
            print(f"Non-numeric yaw in row {i}: yaw={row['header.origin.yaw']}")
            continue  # Skip this row
        
        # Handle missing or invalid yaw
        if pd.isna(yaw):
            yaw = 0  # Default value

        # Translation offsets
        x_header = float(row["header.origin.x"])
        y_header = float(row["header.origin.y"])

        # Rotation matrix (2D only since z is irrelevant)
        R = np.array([
            [np.cos(yaw), -np.sin(yaw)],
            [np.sin(yaw), np.cos(yaw)]
        ])

        # Check for mismatched list lengths
        if len(row["x"]) != len(row["y"]):
            print(f"Row {i} has mismatched lengths:")
            print(f"x: {len(row['x'])}, y: {len(row['y'])}")
            continue

        # Transform each object's coordinates
        x_scs, y_scs = [], []
        for x_vcs, y_vcs in zip(row["x"], row["y"]):
            # Translate
            x_vcs_translated = x_vcs - x_header
            y_vcs_translated = y_vcs - y_header

            # Apply rotation
            vcs_coords = np.array([x_vcs_translated, y_vcs_translated])
            scs_coords = R @ vcs_coords

            # Store transformed coordinates
            x_scs.append(scs_coords[0])
            y_scs.append(scs_coords[1])

        # Append transformed lists to result lists
        x_scs_all.append(x_scs)
        y_scs_all.append(y_scs)

    # Add transformed coordinates back to the dataframe
    processed_labels["x_scs"] = x_scs_all
    processed_labels["y_scs"] = y_scs_all

    return processed_labels


# Transform the coordinates
df_transformed = transform_to_scs_dynamic_list(processed_labels)

# Verify the results
print("Transformed coordinates (SCS):")
print(df_transformed[["x_scs", "y_scs"]])
```

```python
import matplotlib.pyplot as plt

def visualize_vcs_scs(processed_labels, num_cycles=2):
    """
    Visualize the difference between VCS and SCS coordinates for specified cycles.
    Args:
        processed_labels: DataFrame with VCS (`x`, `y`) and SCS (`x_scs`, `y_scs`) columns.
        num_cycles: Number of cycles to visualize.
    """
    for i in range(min(num_cycles, len(processed_labels))):
        row = processed_labels.iloc[i]

        # VCS values
        x_vcs = row["x"]
        y_vcs = row["y"]

        # SCS values
        x_scs = row["x_scs"]
        y_scs = row["y_scs"]

        # Create the scatter plot
        plt.figure(figsize=(8, 6))
        plt.scatter(x_vcs, y_vcs, color="blue", label="VCS", alpha=0.7)
        plt.scatter(x_scs, y_scs, color="red", label="SCS", alpha=0.7)
        
        # Add labels and title
        plt.title(f"Cycle {i + 1}: VCS vs SCS")
        plt.xlabel("X Coordinate")
        plt.ylabel("Y Coordinate")
        plt.legend()
        plt.grid(True)
        
        # Show the plot
        plt.show()

# Visualize the first 2 cycles
visualize_vcs_scs(df_transformed, num_cycles=2)
```

```python
import pandas as pd
import pickle

def process_input_features(file_path, required_columns):
    """
    Process the input features pickle file to retain specific columns
    and convert their values (lists of integers) to lists of floats.
    Args:
        file_path: Path to the pickle file.
        required_columns: List of columns to retain and process.
    Returns:
        Processed DataFrame with only the required columns in numeric format.
    """
    # Step 1: Load the pickle file
    with open(file_path, "rb") as f:
        features_df = pickle.load(f)
    
    # Convert to a pandas DataFrame (if it's not already)
    if not isinstance(features_df, pd.DataFrame):
        features_df = pd.DataFrame(features_df)
    
    # Step 2: Retain only the required columns
    features_df = features_df[required_columns]

    # Step 3: Convert the values to numeric (lists of floats)
    def convert_to_float(value):
        """
        Convert a list of integers (or other types) to a list of floats.
        """
        if isinstance(value, list):
            # Convert each element in the list to float
            return [float(v) for v in value]
        else:
            return value  # If not a list, return as-is (e.g., NaN or other)

    for col in required_columns:
        features_df[col] = features_df[col].apply(convert_to_float)

    return features_df

# Define the file path and required columns
file_path = "D:\\temp\\frr40_detections_0.p"
required_columns = ["angleAzimuth", "angleElevation", "radialVelocity", "distance", "rcs", "radialVelocityDomainMax"]

# Process the input features
processed_features = process_input_features(file_path, required_columns)

# Inspect the processed DataFrame
print("Processed Input Features DataFrame:")
print(processed_features.head())
```

```python
def inspect_radial_velocity_data(features_df):
    """
    Inspect the radial velocity and Nyquist velocity columns in the input features.
    Args:
        features_df: DataFrame containing the input features.
    Returns:
        None. Prints the inspection results.
    """
    # Check the data types of the columns
    print("Data Types:")
    print(features_df[["radialVelocity", "radialVelocityDomainMax"]].dtypes)
    
    # Display a few rows of the data
    print("\nSample Data:")
    print(features_df[["radialVelocity", "radialVelocityDomainMax"]].head())

    # Check for missing or invalid data
    print("\nMissing Values:")
    print(features_df[["radialVelocity", "radialVelocityDomainMax"]].isna().sum())

    # Check the length of lists in radialVelocity
    print("\nLengths of Radial Velocity Lists (First Few Rows):")
    for i, row in features_df.iterrows():
        print(f"Row {i}: {len(row['radialVelocity'])} values")
        if i >= 4:  # Display only the first 5 rows
            break

# Call the function to inspect the data
inspect_radial_velocity_data(processed_features)
```

```python
def unwrap_radial_velocity(features_df):
    """
    Eliminate radial velocity ambiguity for each cycle using the Nyquist velocity.
    Args:
        features_df: DataFrame containing radialVelocity and radialVelocityDomainMax columns.
    Returns:
        Updated DataFrame with a new column trueRadialVelocity containing unwrapped velocities.
    """
    true_radial_velocity_all = []

    for i, row in features_df.iterrows():
        # Extract the measured velocities and the Nyquist velocity
        radial_velocity = row["radialVelocity"]
        nyquist_velocity = row["radialVelocityDomainMax"]
        
        # Initialize the true radial velocity list
        true_radial_velocity = []

        # Process each detected object's radial velocity
        for measured_velocity in radial_velocity:
            # Compute the unwrapped velocity
            unwrapped_velocity = measured_velocity
            while unwrapped_velocity > nyquist_velocity / 2:
                unwrapped_velocity -= nyquist_velocity
            while unwrapped_velocity < -nyquist_velocity / 2:
                unwrapped_velocity += nyquist_velocity
            
            # Store the unwrapped velocity
            true_radial_velocity.append(unwrapped_velocity)

        # Append the result for the current cycle
        true_radial_velocity_all.append(true_radial_velocity)

    # Add the unwrapped velocities as a new column
    features_df["trueRadialVelocity"] = true_radial_velocity_all

    return features_df

# Apply the unwrapping algorithm
processed_features = unwrap_radial_velocity(processed_features)

# Inspect the resulting DataFrame
print("Processed Features with True Radial Velocity:")
print(processed_features[["radialVelocity", "trueRadialVelocity"]].head())
```

```python
import matplotlib.pyplot as plt

def visualize_radial_velocity(features_df, cycle_index=0):
    """
    Visualize the measured radial velocity and the unwrapped true radial velocity for a specific cycle.
    Args:
        features_df: DataFrame containing radialVelocity and trueRadialVelocity columns.
        cycle_index: Index of the cycle to visualize.
    """
    # Extract data for the specified cycle
    measured_velocity = features_df.iloc[cycle_index]["radialVelocity"]
    unwrapped_velocity = features_df.iloc[cycle_index]["trueRadialVelocity"]

    # Number of detected objects in the cycle
    num_objects = len(measured_velocity)

    # Create a line plot for comparison
    plt.figure(figsize=(10, 6))
    plt.plot(range(num_objects), measured_velocity, label="Measured Radial Velocity", marker='o', linestyle='--')
    plt.plot(range(num_objects), unwrapped_velocity, label="Unwrapped Radial Velocity", marker='x', linestyle='-')
    plt.xlabel("Object Index")
    plt.ylabel("Radial Velocity (m/s)")
    plt.title(f"Radial Velocity Comparison for Cycle {cycle_index}")
    plt.legend()
    plt.grid(True)
    plt.show()

# Visualize the radial velocity for the first cycle
visualize_radial_velocity(processed_features, cycle_index=0)
```

```python
def scatter_radial_velocity(features_df, cycle_index=0):
    """
    Scatter plot comparing measured and unwrapped radial velocities for a specific cycle.
    Args:
        features_df: DataFrame containing radialVelocity and trueRadialVelocity columns.
        cycle_index: Index of the cycle to visualize.
    """
    # Extract data for the specified cycle
    measured_velocity = features_df.iloc[cycle_index]["radialVelocity"]
    unwrapped_velocity = features_df.iloc[cycle_index]["trueRadialVelocity"]

    # Create a scatter plot
    plt.figure(figsize=(8, 6))
    plt.scatter(measured_velocity, unwrapped_velocity, alpha=0.7, color="blue")
    plt.plot([min(measured_velocity), max(measured_velocity)],
             [min(measured_velocity), max(measured_velocity)],
             color="red", linestyle="--", label="y = x")
    plt.xlabel("Measured Radial Velocity (m/s)")
    plt.ylabel("Unwrapped Radial Velocity (m/s)")
    plt.title(f"Scatter Plot of Radial Velocity for Cycle {cycle_index}")
    plt.legend()
    plt.grid(True)
    plt.show()

# Visualize the scatter plot for the first cycle
scatter_radial_velocity(processed_features, cycle_index=0)
```

```python
def validate_unwrapping(features_df):
    """
    Validate the unwrapped radial velocity by calculating error statistics.
    Args:
        features_df: DataFrame containing radialVelocity and trueRadialVelocity columns.
    """
    errors = []

    for i, row in features_df.iterrows():
        measured_velocity = row["radialVelocity"]
        true_velocity = row["trueRadialVelocity"]

        # Calculate the absolute error for each object in the cycle
        error = [abs(m - t) for m, t in zip(measured_velocity, true_velocity)]
        errors.extend(error)

    # Calculate summary statistics
    mean_error = sum(errors) / len(errors)
    max_error = max(errors)

    print(f"Mean Absolute Error: {mean_error:.4f}")
    print(f"Max Absolute Error: {max_error:.4f}")

# Validate the unwrapping process
validate_unwrapping(processed_features)
```
