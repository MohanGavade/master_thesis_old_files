```python
# convert all dtypes from int 64 to float 32

columns_to_process = ['rcs', 'distance', 'angleAzimuth', 'angleElevation', 'radialVelocity', ]

def convert_to_float32(df, columns):
    for column in columns:
        df[column] = df[column].apply(lambda X: np.array(X, dtype=np.float32) if isinstance(X, np.ndarray) else X)
    return df
x_cleaned = convert_to_float32(X, columns_to_process)

for column in columns_to_process:
    print(f"\nData types inside '{column}':")
    print(x_cleaned[column].apply(lambda row: row.dtype if isinstance(row, np.ndarray) else type(row)).unique())
```

```python
def process_features(df, columns):
    sequences = []
    for _, row in df.iterrows():
        combined_features = np.column_stack([row[col] for col in columns])
        sequences.append(combined_features)
    return sequences

x_sequences = process_features(x_cleaned, columns_to_process)

print("\nFirst sequence (X):")
print(x_sequences[0])
print(f"shape of first sequence: {x_sequences[0].shape}")
```

```python
import torch
from torch.nn.utils.rnn import pad_sequence

def pad_sequences(sequences):
    tensor_sequences = [torch.tensor(seq, dtype=torch.float32) for seq in sequences]
    padded_sequences = pad_sequence(tensor_sequences, batch_first=True)
    return padded_sequences

x_padded = pad_sequences(x_sequences)

print(f"\nShape of x_padded: {x_padded.shape}")
```

```python
def pad_labels(labels, max_length=None):
    tensor_labels = [torch.tensor(label, dtype=torch.float32) for label in labels]

    if max_length is not None:
        padded_labels = []
        for label in tensor_labels:
            if label.shape[0] < max_length:
                padding = torch.zeros(max_length - label.shape[0], dtype=torch.float32)
                padded_labels.append(torch.cat((label, padding)))
            else:
                padded_labels.append(label[:max_length])
        return torch.stack(padded_labels)
    else:
        return pad_sequence(tensor_labels, batch_first=True)
    

max_length = 272

y_padded = pad_labels(y['orientation_deg'], max_length=max_length)

print(f"\nshape of y_padded: {y_padded.shape}")

```

```python
import random

def mask_features(x, feature_columns, mask_prob=0.2):

    x_masked = x.clone()
    for feature_idx in feature_columns:
        if random.random() < mask_prob:
            x_masked[:, :, feature_idx] = 0
    return x_masked

feature_indices = [0, 1, 2, 3, 4]
x_masked = mask_features(x_padded, feature_indices)

print("sample of masked input features:")
print(x_masked[0])
```

```python
def create_padding_mask(x):

    return (x.sum(dim=2)== 0).int()

padding_mask = create_padding_mask(x_padded)

print("sample padding mask:")
print(padding_mask[0])
```
